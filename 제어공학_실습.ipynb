##**단축키**
*  선택된 셀을 실행: Ctrl+Enter
*  실행 후 다음 셀로 이동: Shift+Enter
*  실행후 다음셀 추가: Alt+Enter  
*  저장: Ctrl+S
*  엔터키: 편집모드
*  ESC: 선택모드
*  마크다운으로 전환: Ctrl+M M
*  코드로 전환: Ctrl+M Y  
*  코드셀에 줄번호 부여: Crtl+M L
*  바로 윗줄에 셀 생성: Ctrl+M A
*  바로 아랫줄에 셀 생성: Ctrl+M B  
*  실행 멈춤: Ctrl+M I

##**타이틀**
 #의 갯수를 통해서 글자 크기를 조절할 수 있다. #의 갯수가 많아질수록 글자의 크기가 작아진다.  
ex)
#손주성
##손주성
###손주성
####손주성
#####손주성

##**강조**  
 *, __, ~~, _ 를 통해서 4가지의 강조 효과를 줄 수 있다.  
Colab에서는 __이 진하게 로 되기 때문에 <,>,u를 사용해야한다.  
ex)  
**손주성**  
<u>손주성</u>  
~~손주성~~  
_손주성_  

##**문장 줄바꿈**  
입력에서 Enter키를 통해 줄바꿈을 하더라도 실제로는 줄바꿈이 되지 않을 수 있다.
줄바꿈을 하고 싶으면 줄바꿈을 하고 싶은 문장 뒤에서 Spacebar을 2번 눌러주면 된다.  
ex)  
손 주 성  
손  
주  
성

##**블럭(인용) 들여쓰기**  

꺽쇠(>)를 이용해서 블럭(인용) 들여쓰기를 할 수 있다.  
ex)  
>손주성  
>>손주성

##**Code Block**  
텍스트 블록에 입력한 부분을 코딩할 때 쓰는 디자인으로 바꿔준다.  
 (`) 3개로 바꾸고 싶은 문장의 앞과 뒤에 두면 그 문장이 바뀐다.

ex)
 ```   
import something
 a = 10
 print(a)
 if a>10:
    print(a)
```

##**체크박스**  
(todo list 같은 것에 좋다) 목록을 표시해 주는 문자 뒤에 대괄호를 통해 체크박스를 만들 수 있으며, 대괄호 안에 x 표시를 넣으면 체크된 체크박스를 만들 수 있다.  
ex)  
 * [ ] 비어있는 체크박스  
 * [x] 체크된 체크박스  

##**하이퍼링크**  
문구로 하이퍼링크를 표시해주고 싶으면 대괄호 안에 문구를 입력 후, 뒤에 괄호안에 링크를 넣어주면 된다.
문구가 필요없다면 꺽쇠를 통해서 바로 하이퍼링크를 만들 수 있다.

ex)  
[바로가기](http://google.com)  
<http://google.com>
